Data Cleaning

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
#convert text to lowercase
df['email'] = df['email'].str.lower()
```

```python
#Tokenisation
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
df['email'] = df['email'].apply(word_tokenize)
```

```python
#remove stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['email'] = df['email'].apply(lambda x: " ".join([word for word in x if word not in stop_words]) )
```

```python
#Lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

df['email'] = df['email'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))

```

```python
#vectorisation
from numpy import vectorize
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['email'])
```
